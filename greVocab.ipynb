```python
import pandas as pd
import numpy as np


from enum import Enum
 
 
class Ebbinghaus(Enum):
    # to review in # minutes
    r1 = 1
    r2 = 5
    r3 = 18
    r4 = 24*60
    r5 = 7*24*60
    r6 = 30*24*60
```

```python
v = pd.read_excel("3000.xlsx", engine='openpyxl', usecols=[8, 13, 14])
v["start_t"] = v.index*3
```

```python
for e in Ebbinghaus:
    v[e.name] = v["start_t"] + e.value * 60
    
```

```python
v
```

```python
to_melt = [e.name for e in Ebbinghaus]
to_melt.append('start_t')
m = pd.melt(v, id_vars=['Word', 'Paraphrase (w/ POS)', 'Paraphrase (English)'], value_vars=to_melt,var_name='rpt', value_name='rpt_tm')
```

```python
v.columns
```

```python
m = m.sort_values('rpt_tm').reset_index(drop = True)
```

```python
m['min'] = m['rpt_tm']/60
m = m.round(0)
m
```

```python
limit = 150

m['day'] = m.index/limit

```

```python
exp = []
i = 0
while (i<m.index.max()):
    exp.append(m.iloc[i:i+limit,:])
    i += limit
```

```python
with open("docs/exp10000.md", "w") as f:
    f.write("### day 1\n")
    f.write(exp[0].to_markdown())
    f.close()
```
